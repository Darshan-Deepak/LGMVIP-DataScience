### IMPORTING LIBRARIES 

```python
#IMPORTING ALL THE ESSENTIAL LIBRARIES FOR OUR PROJECT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### LOADING DATASET

```python
columns = ['sepal length','sepal width','petal length','petal width','class_labels']
#load the data
df = pd.read_csv('E:\DARSHAN\LETSGROWMORE INTERNSHIP\iris.data',names=columns)
df.head()
```

### VISUALIZATION OF DATASET

```python
df.describe()
```

```python
sns.pairplot(df, hue='class_labels')
```

### SEPERATING INPUT AND OUTPUT COLUMNS

```python
data = df.values

X = data[:,0:4]
Y = data[:,4]
```

### SPLITING THE DATA INTO TRAINING AND TESTING

```python
#split the data to train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

### MODEL 1: SUPPORT VECTOR MACHINE ALGORITHM

```python
#support vector machine algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python
prediction1 = model_svc.predict(X_test)
# Caclulating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1)*100)
```

### MODEL2 : LOGISTIC REGRESSION

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
prediction2 = model_LR.predict(X_test)
# Caclulating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction2)*100)
```

### MODEL 3: DECISION TREE CLASSIFIER 

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
prediction3 = model_DTC.predict(X_test)
# Caclulating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction3)*100)
```

```python
# A detailed classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction2)*100)
```
